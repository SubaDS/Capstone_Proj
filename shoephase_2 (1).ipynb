# Join SQL Queries  using all 3 tables:

```python
# Query 1 Write a query finds the top-rated men's shoes along with their sizes from "Table1" and "Table3."
```

```python
query_1 = '''
    SELECT t1.shoe_name, t1.category, t3.size, t3.stars
    FROM table_1_df t1
    INNER JOIN table_3_df t3 ON t1.link = t3.link
    WHERE t1.category = 'M'
    ORDER BY t3.stars DESC
'''
sqldf(query_1, locals())

```

```python
# Query 2 :Write a  query calculates the average comfort rating for each category from "Table1" and "Table3."
```

```python
query = '''
    SELECT t1.category, AVG(t3.comfort) AS avg_comfort_rating
    FROM table_1_df t1
    INNER JOIN table_3_df t3 ON t1.link = t3.link
    GROUP BY t1.category
'''

sqldf(query, locals())
```

```python
# Query 3 : Write a  query identifies products with a durability/quality/performance rating higher than the average from "Table1" and "Table3."
```

```python
query = '''
    SELECT t1.shoe_name, t3.quantification
    FROM table_1_df t1
    INNER JOIN table_3_df t3 ON t1.link = t3.link
    WHERE t3.quantification > (SELECT AVG(quantification) FROM table_3_df)
'''
sqldf(query, locals())
```

```python
# Query 4 : Write a  subquery finds products with comfort ratings above the average comfort rating using "Table1" and "Table3."
```

```python
subquery = '''
    SELECT t1.shoe_name, t3.comfort
    FROM table_1_df t1
    INNER JOIN table_3_df t3 ON t1.link = t3.link
    WHERE t3.comfort > (SELECT AVG(comfort) FROM table_3_df)
'''

query = '''
    SELECT *
    FROM ({}) AS sub
'''.format(subquery)

sqldf(query, locals())
```

```python
# Query 6 :Write a  query that  identifies products with a star rating above the average star rating for their respective size.
```

```python
subquery = '''
    SELECT size, AVG(stars) AS avg_star_rating
    FROM table_3_df
    GROUP BY size
'''

query = '''
    SELECT t1.shoe_name, t3.stars, t3.size
    FROM table_1_df t1
    INNER JOIN table_3_df t3 ON t1.link = t3.link
    INNER JOIN ({}) AS avg_stars ON t3.size = avg_stars.size
    WHERE t3.stars > avg_stars.avg_star_rating
'''.format(subquery)
sqldf(query, locals())
```

```python
# Query 7: Write  a  query finds products with the highest comfort rating in each category.
```

```python
query = '''
    SELECT t1.shoe_name, t3.comfort, t1.category
    FROM table_1_df t1
    INNER JOIN table_3_df t3 ON t1.link = t3.link
    WHERE (t1.category, t3.comfort) IN (
        SELECT t1.category, MAX(t3.comfort)
        FROM table_1_df t1
        INNER JOIN table_3_df t3 ON t1.link = t3.link
        GROUP BY t1.category
    )
'''
sqldf(query, locals())
```
